```python
import pandas as pd


df=pd.read_csv("/content/gdrive/MyDrive/SNA/Results/rankings.csv")

def _intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))


for k in range(1950,2250,250):
    nodes=df[df.columns[1]].to_list()[0:k]
    for i in range(2,14):
        nodes=_intersection(nodes,df[df.columns[i]].to_list()[0:k])
    print(k,"--->",len(nodes),"--->",nodes)
    
```
