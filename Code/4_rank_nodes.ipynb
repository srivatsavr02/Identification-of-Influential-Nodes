```python
import pandas as pd

def _merge(list1, list2):
    merged_list = [(list1[i], list2[i]) for i in range(0, len(list1))]
    return merged_list
df=pd.read_csv("/content/gdrive/MyDrive/SNA/Results/normalized_results.csv")

filept={
    'Rank' : [i for i in range(1,1+len(df['Id'].to_list()))],
}

for i in range(1,13):
    temp=_merge(df['Id'].to_list(),df[df.columns[i]].to_list())
    temp.sort(key=lambda pair: pair[1],reverse=True)
    filept[df.columns[i]]=[pair[0] for pair in temp]
    

newframe=pd.DataFrame(filept)
newframe.to_csv("/content/gdrive/MyDrive/SNA/Results/rankings.csv")
```
