# **Normalise the results of graph measures**


```python
import pandas as pd
from google.colab import drive

drive.mount('/content/gdrive', force_remount=True)

df=pd.read_csv("/content/gdrive/MyDrive/SNA/Results/SBCN-HDP/K-Normalised Results/graph_2200.csv")

filept={
    'Id' : df['Id'],
}
for i in range(1,13):
    temp=df[df.columns[i]].to_list()
    mx=max(temp)
    mn=min(temp)
    result=[(i-mn)/(mx-mn) for i in temp]
    for data in result:
        if data>1 or data<0:
            print("Error")
            exit()
    filept[df.columns[i]]=result

newframe=pd.DataFrame(filept)
newframe.to_csv("/content/gdrive/MyDrive/SNA/Results/SBCN-HDP/K-Normalised Results/graph_2200/graph_2200_normalized.csv")
```

# **Rank the nodes based on centrality measures**

```python
import pandas as pd
from google.colab import drive

drive.mount('/content/gdrive', force_remount=True)

def _merge(list1, list2):
    merged_list = [(list1[i], list2[i]) for i in range(0, len(list1))]
    return merged_list
df=pd.read_csv("/content/gdrive/MyDrive/SNA/Results/SBCN-HDP/K-Normalised Results/graph_2200/graph_2200_normalized.csv")

filept={
    'Rank' : [i for i in range(1,1+len(df['Id'].to_list()))],
}

for i in range(1,13):
    temp=_merge(df['Id'].to_list(),df[df.columns[i]].to_list())
    temp.sort(key=lambda pair: pair[1],reverse=True)
    filept[df.columns[i]]=[pair[0] for pair in temp]
    

newframe=pd.DataFrame(filept)
newframe.to_csv("/content/gdrive/MyDrive/SNA/Results/SBCN-HDP/K-Normalised Results/graph_2200/graph_2200_rankings.csv")
```

# **Find the top k nodes from the ranks**
