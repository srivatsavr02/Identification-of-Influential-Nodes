```python
import pandas as pd
from google.colab import drive
import csv
import networkx as nx

base_path="/content/gdrive/MyDrive/SNA/"
drive.mount('/content/gdrive')

df=pd.read_csv(base_path+"Dataset/SBCNDCTM.csv")
df
```

```python
nodes=set()
edges=[]

for x, y in zip(df['9601108'], df['1001']):
  nodes.add(x)
  nodes.add(y)
  edges.append((x,y))

print("Edges: " + str(len(edges)))
print("Nodes: "+ str(len(nodes)))
```

```python
print(nodes)
```

```python
G = nx.DiGraph()

G.add_nodes_from(nodes)
G.add_edges_from(edges)

print(nx.info(G))
```

# **Degree Centrality = Maximum Possible degree / n-1**

```python
d=nx.degree_centrality(G)
print(d)
k1=[]
kv1= dict()
for key in d:
    k1.append([key,d[key]])
    kv1[key]=d[key]
```

```python
fields=["Id","degree centrality"]
with open(base_path+"Results/degree_centrality.csv",'w') as csvFile:
     csvwriter=csv.writer(csvFile)
     csvwriter.writerow(fields)
     for key in k1:
         csvwriter.writerow(key)
```

```python
from heapq import nlargest
res = nlargest(10, kv1, key = kv1.get)
print("Top 10 Nodes as per degree centrality are: ")
for val in res:
  print(val, kv1[val])
```

# **Degree prestige = (number of incoming links to a node) / (total possible number of incoming links)**

```python
n=len(nodes)

i = dict((v,len(G.edges(v))/(n-1)) for v in G.nodes())

k7=[]
for key in i:
    k7.append([key,i[key]])

fields=["Id","degree_prestige"]
with open(base_path+"Results/degree_prestige.csv",'w') as csvFile:
     csvwriter=csv.writer(csvFile)
     csvwriter.writerow(fields)
     for key in k7:
         csvwriter.writerow(key)
```

# **Proximity prestige = (proportion of vertices in the influence domain) / (mean distance from the vertices)**

```python
distance = []

for dest in G.nodes:
    temp_dis = 0
    n = 0
    for src in G.nodes:
        if (nx.has_path(G,src,dest) == True):
            temp_dis = temp_dis + nx.shortest_path_length(G,source = src,target = dest)
            n = n + 1
    if temp_dis == 0:
        distance.append([dest, 0])
    else:
        distance.append([dest, temp_dis/(n - 1)])

distance

fields=["Id","Proximity-Prestige"]
with open(base_path+"Results/proximity_prestige.csv",'w') as csvFile:
     csvwriter=csv.writer(csvFile)
     csvwriter.writerow(fields)
     for key in distance:
         csvwriter.writerow(key)
```

#**Rank Prestige**

```python
n=len(nodes)

import numpy as np
prominance = np.random.randint(1, 4, size=n)
print("\nASSUME PROMINANCE :\n")
print(prominance)
```

```python
rank_prestige = np.zeros([n], dtype = int)
path_matrix = np.zeros([n, n], dtype = int)

i = 0
j = 0
for src in G.nodes:
    for dest in G.nodes:
        if G.has_edge(dest, src):
            path_matrix[i][j] = 1
        j = j+1
    j = 0
    i = i+1

link=[]
for i in range(n):
    pr_i = 0
    for j in range(n):
        pr_i = pr_i + path_matrix[i][j] * prominance[j]
    rank_prestige[i] = pr_i
    link.append([i,rank_prestige[i]])

print(len(rank_prestige))
```

```python
fields=["Id","rank-Prestige"]
with open(base_path+"Results/rank_Prestige.csv",'w') as csvFile:
     csvwriter=csv.writer(csvFile)
     csvwriter.writerow(fields)
     for key in link:
         csvwriter.writerow(key)
```

# **Percolation states of nodes are used to depict network percolation scenarios over time. In this measure usually the percolation state is expressed as a decimal between 0.0 and 1.0.**

```python
from networkx.algorithms.centrality.percolation import percolation_centrality
nx.set_node_attributes(G,0.1,'percolation')

value=percolation_centrality(G,attribute='percolation')

k=[]
for key in value:
    k.append([key,value[key]])

fields=["Id","Percolation-Centrality"]
with open(base_path+"Results/percolation_centrality.csv",'w') as csvFile:
     csvwriter=csv.writer(csvFile)
     csvwriter.writerow(fields)
     for key in k:
         csvwriter.writerow(key)

print("Top 10 percolation centralities: ")
res = nlargest(10, value, key = value.get)
for val in res:
  print(val, value[val])
```
