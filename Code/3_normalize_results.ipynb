```python
import pandas as pd
from google.colab import drive

drive.mount('/content/gdrive', force_remount=True)

df=pd.read_csv("/content/gdrive/MyDrive/SNA/Results/raw_results.csv")

filept={
    'Id' : df['Id'],
}
for i in range(1,13):
    temp=df[df.columns[i]].to_list()
    mx=max(temp)
    mn=min(temp)
    result=[(i-mn)/(mx-mn) for i in temp]
    for data in result:
        if data>1 or data<0:
            print("Error")
            exit()
    filept[df.columns[i]]=result

newframe=pd.DataFrame(filept)
newframe.to_csv("/content/gdrive/MyDrive/SNA/Results/normalized_results.csv")
```
